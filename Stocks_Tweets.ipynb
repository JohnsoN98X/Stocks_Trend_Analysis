```python
!pip install datasets
from datasets import load_dataset
import pandas as pd
# Downloading dataset
dataset = load_dataset("emad12/stock_tweets_sentiment")
```

```python
# Creating a train datafame
train_data = dataset['train']
train_df = pd.DataFrame(train_data)
train_df.head()
```

```python
test_dataset = dataset['test']
test_df = pd.DataFrame(test_dataset)
test_df.head()
```

```python
# Getting rid of unnecessary columns:
train_df = train_df.drop(columns=['Unnamed: 0', '__index_level_0__', 'input_ids', 'token_type_ids',	'attention_mask'])
```

```python
test_df = test_df.drop(columns=['Unnamed: 0', '__index_level_0__', 'input_ids', 'token_type_ids',	'attention_mask'])
```

```python
missing_ticker_sentences[:20]
```

```python
# As we can see, many of these tweets refers to stocks. Within this, in order to deal with the missing values, two alternatives are considered:
# 1) Writing a loop that will find the authorized stock mentioned in the tweet and enter it in the ticker column.
#    Advantages:
#    * May improve model performance
#    Disadvantages:
#    * Not necessarily the first stock that is mentioned is important.
# 2) Enter a special label for replacing missing values.
#    Advantages:
#    * Maintaining the original information
#    Disadvantages:
#    * Potential for data loss

# Since this is not a very significant part of the data, I will choose at this stage to fill in the missing values ​​with a unique value: 'OTHER'
```

```python
train_df['ticker_symbol'] = train_df['ticker_symbol'].fillna('OTHER')
test_df['ticker_symbol'] = test_df['ticker_symbol'].fillna('OTHER')
train_df.head()
```

```python
# Splitting the train_df into train and validation sets:
from sklearn.model_selection import train_test_split
train_data, val_data = train_test_split(train_df,
                                        random_state=42,
                                        test_size=0.2)
```

```python
# Preparing the labels for models fitting:
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
oh_encoder = OneHotEncoder(sparse_output=False)
label_encoder = LabelEncoder()

# Creating labels variables
train_labels = np.array(train_data['sentiment']).reshape(-1,1)
val_labels = np.array(val_data['sentiment']).reshape(-1,1)
test_labels = np.array(test_df['sentiment']).reshape(-1,1)

# One-Hot-Encoding the labels
train_labels_one_hot = oh_encoder.fit_transform(train_labels)
val_labels_one_hot = oh_encoder.transform(val_labels)
test_labels_one_hot = oh_encoder.transform(test_labels)

# Categoricaly encoding the labels
train_labels_encoded = label_encoder.fit_transform(train_labels)
val_labels_encoded = label_encoder.transform(val_labels)
test_labels_encoded = label_encoder.transform(test_labels)
```

```python
# checking the encoded labels:
print(train_labels_one_hot[:5])
print(train_labels_encoded)
```

```python
# Creating tweets variables:
train_tweets = list(train_data['tweet_cleaned'])
val_tweets = list(val_data['tweet_cleaned'])
test_tweets = list(test_df['tweet_cleaned'])
```

```python
# Creating a function the evaluate the different models:
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score

def evaluate_model(y_true, y_pred):
    results = {
        'Precision_score:': precision_score(y_true, y_pred, average='macro'),
        'Recall_score:': recall_score(y_true, y_pred, average='macro'),
        'F1_score:': f1_score(y_true, y_pred, average='macro'),
        'Accuracy_score:': accuracy_score(y_true, y_pred)
    }
    return results
```

```python
tf.keras.backend.clear_session()
```

```python
# Creating token-level + char-level model:

# 1. Token level model:
tokens_inputs = layers.Input(shape=(1,), dtype=tf.string, name='tokens_level_input')
tokens_vectors = text_vectorizer(tokens_inputs)
tokens_embeds = text_embedding(tokens_vectors)
x = layers.Bidirectional(layers.LSTM(64, return_sequences=True))(tokens_embeds)
x= layers.LSTM(64)(x)
tokens_outputs = layers.Dense(3, activation='softmax', name='tokens_level_output')(x)
tokens_model = tf.keras.Model(tokens_inputs, tokens_outputs, name='tokens_level_model')

# 2. Creating a character-level model:
chars_inputs = layers.Input(shape=(1,), dtype=tf.string, name='char_level_inputs')
char_vectors = char_vectorizer(chars_inputs)
char_embeds = char_embedding(char_vectors)
x = layers.Bidirectional(layers.LSTM(64, return_sequences=True))(char_embeds)
x = layers.LSTM(64)(x)
chars_outputs = layers.Dense(3, activation='softmax', name='char_level_outputs')(x)
chars_model = tf.keras.Model(chars_inputs, chars_outputs, name='chars_level_model')

# 3. Concatenating the Token-Level model and the Character-Level model:
concat_layer = layers.Concatenate(name='concat_layer')([tokens_model.output, chars_model.output])
x = layers.Dense(128, activation='relu', name='dense_128')(concat_layer)
x = layers.Dropout(0.2, name='dropout_layer')(x)
x = layers.Dense(64, activation='relu', name='dense_64')(x)
final_output = layers.Dense(3, activation='softmax', name='final_output')(x)

# 4. Setting the entire model:
model2 = tf.keras.Model(inputs=[tokens_model.input, chars_model.input],
                        outputs=final_output,
                        name='tokens_chars_model')
```

## Model 3 : Tokens Embeddings + Ticker Embeddings
Even though the tickers are categorical values, I'll be using tokenization and embedding:
1. Single-hot coding will produce a large matrix that will cost a lot of computation time.
2. Embedding, in contrast to single-hot coding, can be used for learning different relationships between stocks.


```python
# Creating tickers variables:
train_tickers = list(train_data['ticker_symbol'])
val_tickers = list(val_data['ticker_symbol'])
test_tickers = list(test_df['ticker_symbol'])
```

```python
# Some of the tickers contain more than one word only. In order to treat each ticker as a single entity, I will replace the spaces with an underscore:
def fix_ticker (tickers):
  tickers = [ticker.replace(" ", "_") for ticker in tickers]
  return tickers
```

```python
train_tickers = fix_ticker(train_tickers)
val_ticker = fix_ticker(val_tickers)
test_tickers = fix_ticker(test_tickers)
```

```python
# Checking the number of tickers:
n_unique_tickers = len(np.unique(train_tickers))
n_unique_tickers
```

```python
# Creating a ticker vectorizer layer:
ticker_vectorizer = layers.TextVectorization(max_tokens=n_unique_tickers)
```

```python
# Adappting the ticker vectorizer:
ticker_vectorizer.adapt(train_tickers)
```

```python
# Checking the performance of the ticker vectorizer:
import random
random_ticker = random.choice(np.unique(train_tickers))
print(f'Random ticker: {random_ticker}')
print(f'Vectorized ticker: {ticker_vectorizer(random_ticker)}')
```

```python
# Creating a tiker embedding layer:
ticker_embedding = layers.Embedding(input_dim=n_unique_tickers,
                                   output_dim=16,
                                   mask_zero=False)
```

```python
# Checking the performance of the tiker embedding:
random_ticker = random.choice(np.unique(train_tickers))
print(f'Random ticker: {random_ticker}')
print(f'Vectorized ticker: {ticker_vectorizer(random_ticker)}')
print(f'Embedded ticker:\n {ticker_embedding(ticker_vectorizer(random_ticker))}')
print(f'Embeddd ticker shape: {ticker_embedding(ticker_vectorizer(random_ticker)).shape}')
```

```python
# Creating a model

# 1) Creating token-level model:
text_inputs = layers.Input(shape=(1,), dtype=tf.string, name='text_level_inputs')
text_vectors = text_vectorizer(text_inputs)
text_embeds = text_embedding(text_vectors)
x = layers.Bidirectional(layers.LSTM(64, return_sequences=True))(text_embeds)
x = layers.LSTM(128)(x)
text_outputs = layers.Dense(3, activation='softmax', name='text_level_outputs')(x)
text_model = tf.keras.Model(text_inputs, text_outputs, name='text_level_model')

# 2) Creating tiker model:
ticker_inputs = layers.Input(shape=(1,), dtype=tf.string, name='ticker_lv_inputs')
ticker_vectors = ticker_vectorizer(ticker_inputs)
ticker_embeds = ticker_embedding(ticker_vectors)
ticker_pool = layers.GlobalAveragePooling1D()(ticker_embeds)
ticker_dense = layers.Dense(128, activation='relu')(ticker_pool)
ticker_output = layers.Dense(3, activation='softmax', name='ticker_level_outputs')(ticker_dense)
ticker_model = tf.keras.Model(ticker_inputs, ticker_output, name='ticker_level_model')

# 3) Concatinating the token-level model and the ticker model:
text_chars_layer = layers.Concatenate(name='concat_layer')([text_model.output, ticker_model.output])
x = layers.Dense(128, activation='relu')(text_chars_layer)
x = layers.Dropout(0.3)(x)
x = layers.Dense(64, activation='relu')(x)
final_outputs = layers.Dense(3, activation='softmax', name='final_output_layer')(x)

# 4) Setting the final model:
model3 = tf.keras.Model(inputs=[text_model.input, ticker_model.input],
                        outputs = final_outputs,
                        name='tweets_tickers_model')
```

```python
model3.summary()
```

```python
plot_model(model3, show_shapes=True, show_layer_names=True)
```

```python
# Compile model3:
from tensorflow.keras.optimizers import Adam
model3.compile(loss='categorical_crossentropy',
               optimizer=Adam(clipnorm=1, learning_rate=0.0001),
               metrics=['accuracy'])
```

```python
# Creating dataset to fit the model
train_dataset = tf.data.Dataset.from_tensor_slices(((train_tweets, train_tickers), train_labels_one_hot))
val_dataset = tf.data.Dataset.from_tensor_slices(((val_tweets, val_tickers), val_labels_one_hot))
test_dataset = tf.data.Dataset.from_tensor_slices(((test_tweets, test_tickers), test_labels_one_hot))
```

```python
train_dataset = train_dataset.batch(16).prefetch(tf.data.AUTOTUNE)
val_dataset = val_dataset.batch(16).prefetch(tf.data.AUTOTUNE)
test_dataset = test_dataset.batch(16).prefetch(tf.data.AUTOTUNE)
```

```python
# fit the model:
hist3 = model3.fit(train_dataset,
                   steps_per_epoch = int(0.1*len(train_dataset)),
                   epochs=3,
                   validation_data = val_dataset,
                   validation_steps = int(0.1*len(val_dataset))
                   )
```

```python
# Making predictions with model3:
model3_pred_probs = model3.predict(val_dataset)
model3_pred = tf.argmax(model3_pred_probs, axis=1)
```

```python
# Creating a model:

# 1) Creating token-level model:
text_inputs = layers.Input(shape=(1,), dtype=tf.string, name='text_lvl_inputs')
text_vectors = text_vectorizer(text_inputs)
text_embeds = text_embedding(text_vectors)
text_bidi = layers.Bidirectional(layers.LSTM(64, return_sequences=True))(text_embeds)
text_dropout = layers.Dropout(0.5)(text_bidi)
text_lstm = layers.LSTM(128)(text_dropout)
text_outputs = layers.Dense(3, activation='softmax', name='text_lvl_outputs')(text_lstm)
text_model = tf.keras.Model(text_inputs, text_outputs, name='txt_model')

# 2) Creating tiker model:
ticker_inputs = layers.Input(shape=(1,), dtype=tf.string)
ticker_vectors = ticker_vectorizer(ticker_inputs)
ticker_embeds = ticker_embedding(ticker_vectors)
ticker_pool = layers.GlobalAveragePooling1D(name='ticker___pool')(ticker_embeds)
ticker_dense = layers.Dense(128, activation='relu')(ticker_pool)
ticker_dropout = layers.Dropout(0.5)(ticker_dense)
ticker_output = layers.Dense(3, activation='softmax', name='ticker_level_outputs')(ticker_dropout)
ticker_model = tf.keras.Model(ticker_inputs, ticker_output, name='ticker_level_model')

# 3) Creating a tweet-length model:
length_inputs = layers.Input(shape=(1,), dtype=tf.float64, name='tweets_lengths_input')
length_dense = layers.Dense(16, activation='relu')(length_inputs)

# 3) Concatinating the token-level model and the ticker model:
text_chars_layer = layers.Concatenate(name='concat_layer')([text_model.output, ticker_model.output, length_dense])
x = layers.Dense(128, activation='relu')(text_chars_layer)
x = layers.Dropout(0.5)(x)
x = layers.Dense(64, activation='relu')(x)
final_outputs = layers.Dense(3, activation='softmax', name='final_output_layer')(x)

# 4) Setting the final model:
model4 = tf.keras.Model(inputs=[text_model.input, ticker_model.input, length_inputs],
                        outputs = final_outputs,
                        name='tweets_tickers_lengths_model')
```
